```python
import numpy as np
import matplotlib.pyplot as plt

# Factor de lorentz:  y(V)=1/np.sqrt(1-v**2/c**2)
c=1
v = np.linspace(0, 0.99, 500)
y= 1/np.sqrt(1-v**2/c**2)

plt.figure(figsize=(8, 6))
plt.plot(v, y, label=r'$\gamma$')
plt.xlabel(r'$v/c$', fontsize=14)
plt.ylabel(r'$\gamma$', fontsize=14)
plt.legend()
plt.axvline(x=0.9, color='red', linestyle='--')
plt.axhline(y=2.29, color='red', linestyle='--')
plt.text(0.91, 2.3, 'v = 0.9c', color='red')
plt.axvline(x=0.1, color='red', linestyle='--')
plt.axhline(y=1, color='red', linestyle='--')
plt.text(0.1, 1, 'v = 0.1c', color='red')

plt.show()
```
